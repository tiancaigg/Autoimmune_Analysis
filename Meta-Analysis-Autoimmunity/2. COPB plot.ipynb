```r
library(tidyverse)
```

```r
Res_merged <- readRDS("generated_data/Res_merged.rds")
```

```r
Res_merged %>% pull(source) %>% unique()
```

#### change MEcyan to MEruby according to the my manuscript

```r
Res_merged <- Res_merged %>%
  mutate(across(c(module, Module_n), 
                ~gsub("MEcyan", "MEruby", .)))
```

```r
Res_merged
```

```r
Res_merged_two_module <-  Res_merged %>% filter(module %>% str_detect("MEruby|MEblue")) %>% group_by(source, module) %>% summarise(count = n(), mean_l2fc = mean(log2FC), median_l2fc = median(log2FC)) %>% ungroup
```

```r
Res_merged_two_module 
```

```r
Res_merged_two_module %>% sample_n(10)
```

### MEruby is immunity, MEblue is growth

```r
width_ = 7.7
height_ = 6.3
options(repr.plot.width= width_ , repr.plot.height= height_ , repr.plot.res = 200)

# Reshape the data for plotting
plot_data <- Res_merged_two_module %>%
  select(source, module, median_l2fc) %>%
  pivot_wider(names_from = module, values_from = median_l2fc) %>%
  filter(!is.na(MEruby) & !is.na(MEblue))

# Create the dot plot
ggplot(plot_data, aes(x = MEruby, y = MEblue)) +
  geom_point(size = 3, alpha = 0.7) +
  geom_text(aes(label = source), vjust = -0.8, hjust = 0.5, size = 3) +
  theme_bw() +
  labs(
    title = "Median log2 fold change comparison",
    x = "MEruby module (median_l2fc)",
    y = "MEblue module (median_l2fc)"
  ) +
  geom_hline(yintercept = 0, linetype = "dashed", color = "darkgray") + 
  geom_vline(xintercept = 0, linetype = "dashed", color = "darkgray")
```
