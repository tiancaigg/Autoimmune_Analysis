```r
library(tidyverse)

#####################
##  GENERAL SETUP  ##
#####################

# Keep the same setup code for functions
theme_set(theme_bw(base_family = "Arial", base_size = 10))
update_geom_defaults("text", list(size = 8/.pt))

# Helper function to join path components
mkpath <- function(...){paste(..., sep = '/')}

# Use current working directory instead of hardcoded path
dir_proj <- getwd()
cat("Using current working directory:", dir_proj, "\n")

# Create the plots directory if it doesn't exist
dir.create(file.path(dir_proj, "plots"), showWarnings = FALSE, recursive = TRUE)

# Define the save_plot function to save to plots/ directory
save_plot <- function(fname, p, w = 9.91, h = 5.44, units = "in", fmt = c("pdf", "png")) {
  # Create plots directory if it doesn't exist
  plot_dir <- file.path(dir_proj, "plots")
  dir.create(plot_dir, showWarnings = FALSE)
  
  # Save in each specified format
  for (ext in fmt) {
    filename <- file.path(plot_dir, paste0(fname, ".", ext))
    if (ext == "pdf") {
      ggsave(filename, plot = p, width = w, height = h, units = units, device = cairo_pdf)
    } else {
      ggsave(filename, plot = p, width = w, height = h, units = units)
    }
    cat("Saved plot to", filename, "\n")
  }
}

#####################
## LOAD DATA       ##
#####################

# Assuming raw_rds is already loaded in your environment
# If not, you'll need to load it:
# raw_rds <- readRDS("path/to/Res_merged.rds")

df.wgcna <- Res_merged
df.wgcna.median <- df.wgcna %>%
    dplyr::group_by(source, module) %>%
    dplyr::summarise(l2fc.median = median(log2FC, na.rm = TRUE))
df.wgcna.median <- df.wgcna.median %>%
    dplyr::mutate(immune.type = as.factor(case_when(
                      str_detect(source, "Rpt|Rpm|Rps") ~ "infection",
                      TRUE ~ "autoimmune"
                  )))

# Create the dataframe for plotting
df.toplot <- df.wgcna.median %>%
    tidyr::spread(module, l2fc.median) %>%
    # Instead of storing image paths, we'll use shape and color to differentiate
    # We'll still keep track of which would have had images
    dplyr::mutate(
        has_image = str_replace(source, '/', '-') %in% 
                  list.files(paste0(dir_proj, "/images/drawn"), pattern = ".png"),
        # Create a label field for plotting
        plot_label = source
    )
```

```r
#####################
## VECTOR PLOTS    ##
#####################
width_ <- 5
height_ <- 5
options(repr.plot.width = width_, repr.plot.height = height_, repr.plot.res = 200)

# # Version with labels
# p.lab <- p.nolab + 
#     ggrepel::geom_text_repel(aes(label = source), size = 2) +
#       theme_Publication() +
#     theme(axis.text = element_text(size = 8),
#           axis.title = element_text(size = 8),
#           # axis.title.y = element_blank()
#          ) +
#     theme(text = element_text(size = 8, family = "Arial"), 
#           strip.text = element_text(size = 12))

# # p.lab
# save_plot("fig1_v2_lab", p.lab, h = height_, w = width_)
# cat("Plots saved to:", file.path(dir_proj, "plots"), "\n")

# For the ADR1 KO vector plot
df.tmp <- df.toplot %>%
    dplyr::filter(str_detect(source, "adr1s")) %>%
    dplyr::mutate(reference = str_extract(source, ".+?(?=_adr1s)"))
df.toplot.vector.adr <- df.toplot %>%
    dplyr::left_join(df.tmp %>% dplyr::rename(adr1ko = source), by = c("source" = "reference")) %>%
    dplyr::filter(as.character(source) %in% c(df.tmp$reference, as.character(df.tmp$source)))

p.adr1 <- p.base +
    # Background points
    geom_point(data = df.toplot, aes(y = MEruby, x = MEblue, fill = immune.type),
               colour = "black", pch = 21, alpha = 0.3) +
    # Highlighted points
    geom_point(data = df.toplot.vector.adr,
               aes(y = MEruby.x, x = MEblue.x, fill = immune.type.x),
               colour = "black", pch = 21, size = 3, show.legend = FALSE) +
    # Arrows between points
    geom_segment(data = df.toplot.vector.adr,
                 aes(y = MEruby.x, x = MEblue.x, yend = MEruby.y, xend = MEblue.y),
                 lineend = "butt", linejoin = "round",
                 arrow = arrow(length = unit(0.02, "npc"), type = "closed")) +
    # Origin point
    geom_point(x = 0, y = 0, size = 3, pch = 21, colour = "black", fill = NA) +
    # Special segment
    geom_segment(data = df.toplot.vector.adr %>% dplyr::filter(source == "Col-0_adr1s"),
                 aes(y = 0, x = 0, yend = MEruby.x, xend = MEblue.x),
                 lineend = "butt", linejoin = "round",
                 arrow = arrow(length = unit(0.02, "npc"), type = "closed"),
                 lty = 2) +
    # Legend
    guides(fill = guide_legend(override.aes = list(alpha = 1, size = 3))) +
    # Labels
    geom_text(data = df.toplot.vector.adr %>% dplyr::filter(!str_detect(source, "adr1")),
              aes(y = MEruby.x, x = MEblue.x, label = source),
              size = 3, nudge_y = -0.15, nudge_x = -0.2) +
    geom_text(data = data.frame(lab = "Col-0", x = 0, y = 0),
              aes(y = y, x = x, label = lab),
              size = 3, nudge_y = -0.15, nudge_x = -0.2)

# Time course vector plot - no changes needed as it doesn't use ggimage
# This part can remain essentially the same as in your original code

# Save the vector plots
save_plot("vector_adr1", p.adr1, h = 4, w = 4)
p.adr1
# If you're also creating the pathogen time course plot (p.time), uncomment this:
# df.tmp <- df.toplot %>%
#     dplyr::filter(immune.type == "infection") %>%
#     [...rest of time course code...]
# p.time <- p.base + [...rest of time course plot code...]
# save_plot("vector_pathogen", p.time, h = 4, w = 4)

cat("All plots saved to:", file.path(dir_proj, "plots"), "\n")
```
