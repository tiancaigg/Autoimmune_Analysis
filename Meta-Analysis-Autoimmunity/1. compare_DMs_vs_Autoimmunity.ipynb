```r
merged_l2FC_all <- merged_l2FC_clean %>% dplyr::select(-gene_name) %>% dplyr::rename(source = condition) %>% 
    dplyr::rename(AGI = gene_id) %>%
    rbind(res_DMs)   %>% 
    rbind(res_Saile_Mine %>% filter(source %>% str_detect("Saile")) %>% mutate(source = source %>% str_replace("_vs_Col.T0.NA", '')) %>% dplyr::rename(AGI = gene_id) ) %>%
    left_join(map_id_name, by = c("AGI" = "gene_id"))
```

```r
# Modified analysis with ordered factors
NLRs_and_TIRs %>% 
  filter(Class != 'TIR') %>% 
  left_join(merged_l2FC_all, by = c("AGI")) %>%
  filter(log2FC > 2) %>%
  group_by(source) %>% 
  summarise(NLRs_up = n()) %>%
  # Convert source to factor with custom order
  mutate(source = factor(source, levels = levels(combined_l2FC_all$source))) %>%
  # Arrange by the new factor levels
  arrange(source)
```

```r
up_NLRs_table <- NLRs_and_TIRs %>%
  filter(Class != 'TIR') %>%
  left_join(merged_l2FC_all, by = c("AGI")) %>%
  group_by(source) %>%
  summarise(
    detected_NLR = sum(!is.na(log2FC)),
    NLRs_up = sum(log2FC > 2, na.rm = TRUE),
    .groups = 'drop'
  ) %>%
  arrange(source)
up_NLRs_table
up_NLRs_table %>% write_csv("generated_data/up_NLRs_table.csv")
```

```r
# Create wide format table with NLRs and their log2FC values across sources
NLRs_and_TIRs %>%
  filter(Class != 'TIR') %>%
  left_join(merged_l2FC_all, by = "AGI") %>%
  select(AGI, Class, Symbol, source, log2FC) %>%
  pivot_wider(
    id_cols = c(AGI, Class, Symbol),
    names_from = source,
    values_from = log2FC
  ) %>%
  arrange(AGI)
```

```r
# Create wide format table with NLRs and their log2FC values across sources
all_NLRs_and_TIRs_across_sources_log2FC <- NLRs_and_TIRs %>%
  left_join(merged_l2FC_all, by = "AGI") %>%
  select(AGI, Class, Symbol, source, log2FC) %>%
  pivot_wider(
    id_cols = c(AGI, Class, Symbol),
    names_from = source,
    values_from = log2FC
  ) %>%
  arrange(AGI) 
all_NLRs_and_TIRs_across_sources_log2FC
all_NLRs_and_TIRs_across_sources_log2FC %>% write_csv("generated_data/NLRs_and_TIRs_log2FC.csv")
```
