```r
library(data.table)

base_dir <- "/media/HDD2/donghui/bulk_ATAC_DM1_DM2d/TOBIAS/BINDetect_output_v3/"
tf_dirs <- list.dirs(path = base_dir, full.names = TRUE, recursive = FALSE)

read_overview_data <- function(tf_dir) {
  overview_file <- file.path(tf_dir, paste0(basename(tf_dir), "_overview.txt"))
  
  if (file.exists(overview_file)) {
    data <- fread(overview_file)
    data[, TF := basename(tf_dir)]  # Add a column for the TF
    return(data)
  } else {
    return(data.table())
  }
}

# Apply the function to each TF directory
all_overview_data <- lapply(tf_dirs, read_overview_data)

# Combine all data into one dataframe
merged_overview_data <- rbindlist(all_overview_data, fill = TRUE)
merged_overview_data_bound_in_DM_or_COL <- merged_overview_data %>% filter( COL_bound == '1' | DM_bound == '1')
```

```r
library(dplyr)
library(clusterProfiler)
library(org.At.tair.db)
library(parallel)  # for parallel processing
```

```r
# Assuming DM_unique_bound_peaks$gene_id contains the unique TAIR IDs
TF_targets <- list(
  COL_unique = unique(COL_unique_bound_peaks$gene_id),
     DM_unique = unique(DM_unique_bound_peaks$gene_id)
)
TF_targets
# Define the function for enrichment
enrich_func <- function(genes) {
  enrichGO(
    gene = genes,
    OrgDb = org.At.tair.db,
    keyType = "TAIR",
    ont = "BP",
    pAdjustMethod = "BH",
    qvalueCutoff = 0.05
  )
}

# Perform enrichment analysis (consider the available cores in your system)
ego_list <- mclapply(TF_targets, enrich_func, mc.cores = min(3, detectCores()))

# Name the list according to your gene lists
names(ego_list) <- names(TF_targets)

# Function to simplify the results with a given cutoff
simplify_ego_list <- function(ego_list, cutoff) {
  lapply(ego_list, function(ego) {
    if (!is.null(ego)) simplify(ego, cutoff = cutoff)
    else NULL
  })
}

# Apply the simplification with a cutoff of your choice
ego_list_simplified <- simplify_ego_list(ego_list, cutoff = 0.9)
```
