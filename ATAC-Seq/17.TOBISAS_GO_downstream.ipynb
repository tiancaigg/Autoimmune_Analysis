```r
library(dplyr)
library(clusterProfiler)
library(org.At.tair.db)
library(parallel)  # for parallel processing

# Function to perform TF analysis
perform_tf_analysis <- function(tf_name) {
  # Define file paths
  base_dir <- "/media/HDD2/donghui/bulk_ATAC_DM1_DM2d/TOBIAS/BINDetect_output_v3/"
  DM_bound_file <- paste0(base_dir, tf_name, "/beds/", tf_name, "_DM_bound.bed")
  COL_bound_file <- paste0(base_dir, tf_name, "/beds/", tf_name, "_COL_bound.bed")

  # Read the files
  DM_bound <- read.csv(DM_bound_file, sep = '\t', header = F)
  COL_bound <- read.csv(COL_bound_file, sep = '\t', header = F)

  # Intersection (Common rows)
  common_rows <- inner_join(DM_bound, COL_bound, by = c("V1", "V2", "V3"))

  # Difference (Rows in DM_bound not in COL_bound)
  unique_DM <- anti_join(DM_bound, COL_bound, by = c("V1", "V2", "V3"))

  # Difference (Rows in COL_bound not in DM_bound)
  unique_COL <- anti_join(COL_bound, DM_bound, by = c("V1", "V2", "V3"))

  TF_targets <- list(
    common = common_rows$V13.x %>% unique, 
    DM_unique = unique_DM$V13 %>% unique, 
    COL_unique = unique_COL$V13 %>% unique
  )

  enrich_func <- function(genes) {
    enrichGO(gene = genes,
             OrgDb = org.At.tair.db,
             keyType = "TAIR",
             ont = "BP",
             pAdjustMethod = "BH",
             qvalueCutoff = 0.05)
  }

  # Perform enrichment analysis in parallel
  ego_list <- mclapply(TF_targets, enrich_func, mc.cores = 20)
  names(ego_list) <- names(TF_targets)

  # Function to apply simplify with different cutoffs
  simplify_ego_list <- function(ego_list, cutoff) {
    lapply(ego_list, function(ego) {
      if (!is.null(ego)) simplify(ego, cutoff = cutoff)
      else NULL
    })
  }

  # Simplify the ego list
  simplified_ego_list <- simplify_ego_list(ego_list, cutoff = 0.4)

  # Convert to dataframe
  convert_to_dataframe <- function(ego_list) {
    combined_df <- data.frame()
    for (name in names(ego_list)) {
      ego <- ego_list[[name]]
      if (!is.null(ego)) {
        df <- as.data.frame(ego)
        if (nrow(df) > 0) {
          df$Source <- name
          combined_df <- rbind(combined_df, df)
        }
      }
    }
    return(combined_df)
  }

  # Convert your simplified_ego_list to a dataframe
  simply_GO <- convert_to_dataframe(simplified_ego_list) %>% mutate( tf_name = tf_name)
  simply_GO %>% write.csv(paste0('TF_GO_module/simply_GO_', tf_name ,'.csv'), row.names = FALSE, quote = FALSE)
  return(simply_GO)
}

# Function to get the TF name from a directory
get_tf_name <- function(base_dir, tf_id) {
  dirs <- list.dirs(base_dir, full.names = FALSE, recursive = FALSE)
  tf_name <- dirs[grep(tf_id, dirs)]
  return(tf_name)
}

sub_dir_name <- get_tf_name(base_dir ="/media/HDD2/donghui/bulk_ATAC_DM1_DM2d/TOBIAS/BINDetect_output_v3/", 'AT3G04060')
perform_tf_analysis(sub_dir_name)
```

```r
library(parallel)

# Modified function to process multiple TFs using parallel processing
process_multiple_TFs_parallel <- function(tf_ids, base_dir) {
  # Function to process a single TF
  process_single_tf <- function(tf_id) {
    sub_dir_name <- get_tf_name(base_dir, tf_id)
    if (length(sub_dir_name) > 0) {
      return(perform_tf_analysis(sub_dir_name))
    } else {
      return(data.frame())  # Return empty dataframe if no directory found
    }
  }

  # Use mclapply for parallel processing
  results_list <- mclapply(tf_ids, process_single_tf, mc.cores = 18)

  # Combine all results into one dataframe
  all_results <- do.call(rbind, results_list)
  return(all_results)
}

# Define TF IDs and base directory
base_dir <- "/media/HDD2/donghui/bulk_ATAC_DM1_DM2d/TOBIAS/BINDetect_output_v3/"

# Process all TFs in parallel and get combined result
combined_result <- process_multiple_TFs_parallel(TFs_of_interest, base_dir)
combined_result
```

```r
combine_data_frames <- function(df_list) {
  # Get all unique column names across all data frames
  all_col_names <- unique(unlist(lapply(df_list, names)))

  # Add missing columns to each data frame
  df_list <- lapply(df_list, function(df) {
    missing_cols <- setdiff(all_col_names, names(df))
    df[missing_cols] <- NA
    return(df)
  })

  # Ensure the same column order for all data frames
  df_list <- lapply(df_list, function(df) df[all_col_names])

  # Combine all data frames
  combined_df <- do.call(rbind, df_list)
  return(combined_df)
}

# Usage example
# Assuming df1, df2, df3 are your data frames
node_combined <- combine_data_frames(list(GO_nodes, TF_nodes)) %>% distinct
node_combined %>% nrow
node_combined
```

```r
# Defining a custom function to find the most possible module
find_most_possible_module <- function(row) {
  # Extracting only the _per columns
  per_values <- row[grep("_per$", names(row))]
  
  # Find the max value and its index
  max_val <- max(per_values, na.rm = TRUE)
  max_index <- which.max(per_values)

  # Check if the max value is greater than 0.4 and unique
  if (max_val > 0.4 && sum(per_values == max_val, na.rm = TRUE) == 1) {
    return(names(per_values)[max_index])
  } else {
    return(NA)
  }
}

# Apply the function to each row of the dataframe
combined_result2_updated$most_possible_module <- apply(combined_result2_updated, 1, find_most_possible_module)

# View the updated dataframe
combined_result2_updated %>%  filter(most_possible_module %>% str_detect('ME')) %>% 
    dplyr::select(-geneID, -pvalue, -BgRatio, -p.adjust, -log2FoldChange_DM1_2d_vs_Col_14, -tf_name)
```
